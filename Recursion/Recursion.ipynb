```python
# 1,2,3 | 1,3,2 | 2,3,1
# permute(1, permute(2,3)) permute(2, permute(3))
def permute(lst):
    if len(lst) == 0:
        return []
    if len(lst) == 1:
        return [lst]
    temp_lst = []
    for i in range(len(lst)):
        item = lst[i]
        remain = lst[:i] + lst[i+1:]
        for p in permute(remain):
            temp_lst.append([item] + p)
    return temp_lst

permute([1,2,4])
```

```python
# powerset [1,2,3], powerset(1, powerset(2,3)) powerset(2, powerset(3))
def powerset(lst):
    if len(lst) == 0 :
        return [[]]
    rest = powerset(lst[1:])
    with_first = [[lst[0]] + p for p in rest]
    without_first = rest
    return with_first + without_first

powerset([1,2,3])
```

```python
# subset size k
def subsetk(temp_lst, temp_idx, _input, input_idx, k):
    if temp_idx == k:
        print(temp_lst)
        return
    if input_idx >= len(_input):
        return
    temp_lst[temp_idx] = _input[input_idx]
    subsetk(temp_lst, temp_idx+1, _input, input_idx+1, k)
    subsetk(temp_lst, temp_idx, _input, input_idx+1, k)
    
subsetk([None, None, None], 0, [1,2,3,4], 0, 3)
```

```python
# filling up th eboard from left to right
def is_safe(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, len(board), 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True
    
def NQueen(board, col):
    if col >= len(board):
        return True
    for row_i in range(len(board)):
        if is_safe(board, row_i, col):
            board[row_i][col] = 1
            if NQueen(board, col+1):
                return True
            board[row_i][col] = 0
    return False

K = 8
board = [[0 for i in range(K)] for j in range(K) ]
NQueen(board, 0)
for row in board:
    print(row)
```

```python
# generate string of matching parenthesis
# ()() (())
# ()()() ()(()) (())() ((())) 

def MParent(num_left, num_right, lst):
#     print(num_left, num_right, lst)
    if num_left == 0 and num_right ==0:
        print(lst)
        return
    if num_left > 0:
        MParent(num_left-1, num_right, lst+'(')
    if num_left< num_right:
        MParent(num_left, num_right-1, lst+')')
        
        
MParent(3,3, '')
```

```python
# Generate Binary Trees
# BT(3) node+BT(2)
import random
class BTNode:
    def __init__(self, val, left, right):
        self.value = val
        self.left = left
        self.right = right
        
def genBT(num_nodes):
    results = [] # added this
    if num_nodes ==0:
        results.append(BTNode(0, None, None))
 
    for i in range(0, num_nodes, 1):
        num_left = i
        num_right = num_nodes-num_left -1
#         print(num_left, num_right)
#         print('Left, genBT({}, {})'.format(num_left, results))
        lefts = genBT(num_left)
#         print('Right, genBT({}, {})'.format(num_right, results))
        rights = genBT(num_right)
        for left in lefts:
            for right in rights:
                results.append(BTNode(0, left, right))

    return results

results = genBT(4)
```

```python
# GrayCode 0, 1 
#         00, 01, 11, 10 
#         000,001, 011, 010, 110,111,101, 100

def graycode(n):
    if n==1:
        return [[0],[1]]
    if n ==0:
        return
    temp = graycode(n-1)
    print(temp)
    add_0 = [[0] + item for item in temp]
    add_1 = [[1]+ temp[i] for i in range(len(temp)-1, -1, -1)]
    return add_0+add_1

graycode(3)
```

```python
def TreeD(node):
    if not node:
        return 0
    
    left_height = 0
    right_height = 0
    if node.left:
        left_hight = 1 + TreeD(node.left)
    if node.right:
        right_height = 1 + TreeD(node.right)
    
def diam(node):
    if not node:
        return 0
    
    left_h = TreeD(node.left)
    right_h = TreeD(node.right)
    left_d = diam(node.left)
    right_d = diam(node.right)
    return max( (left_h + right_h), max(left_d, right_d))
    
```
